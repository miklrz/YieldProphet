```python
import yfinance as yf
from prophet import Prophet
import pandas as pd

# 1. Загрузка данных
# Возьмем, к примеру, индекс S&P 500 (^GSPC) или любую акцию
ticker = "AAPL"
data = yf.download(ticker, start="2020-01-01", end="2024-05-01")
```

```python
data.head()
```

```python
# 2. Преобразование для Prophet
df_prophet = data[["Close"]].reset_index()
df_prophet.columns = ["ds", "y"]

df_prophet["ds"] = df_prophet["ds"].dt.tz_localize(None)
print("DF ready")
df_prophet.head()
```

```python
# 3. Инициализация и обучение модели
model = Prophet(daily_seasonality=True)  # для финансовых данных часто полезно
model.fit(df_prophet)

# 4. Создание датафрейма для будущего прогноза (например, на 30 дней вперед)
future = model.make_future_dataframe(periods=30)
forecast = model.predict(future)

# Посмотрим на результат
print(forecast[["ds", "yhat", "yhat_lower", "yhat_upper"]].tail())
```
